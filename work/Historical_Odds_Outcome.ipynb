```python
import pandas as pd
```

```python
df = pd.read_csv('../data/odds_data.csv')
df = df.dropna()
df
```

```python
df = df.replace({'title_bout': {False:0,True: 1}})
cols = [5,6,10]
df = df.drop(df.columns[cols],axis=1)
df
```

```python
df_dum = pd.get_dummies(df, columns=['weight_class'])
df_dum
```

```python
df_dum = df_dum.replace({'Winner': {'Blue':0,'Red': 1}})
df_dum = df_dum.drop(df_dum[df_dum.Winner == 'Draw'].index)
df_dum = df_dum.astype({'Winner': 'int32'})
df_dum['date'] = pd.to_datetime(df_dum['date'])
df_dum
```

```python
data = pd.read_csv('../data/data.csv')
data
```

```python
col = [4,5,6,7,8]
df2 = data.drop(data.columns[col],axis=1)
df2['date'] = pd.to_datetime(df2['date'])
df2
```

```python
final_df = pd.merge(df_dum,df2, on=['R_fighter','B_fighter','date'])
final_df
```

```python
for column in final_df.columns:
    if final_df[column].isnull().sum()!=0:
        print(f"Nan in {column}: {final_df[column].isnull().sum()}")
```

```python
final_df = final_df.dropna()
```

```python
final_df.drop(columns=['R_fighter','B_fighter','R_Stance','B_Stance','B_draw','R_draw','Referee','date'], inplace=True)
final_df
```

```python
cnt = 0
for col in final_df.columns: 
    cnt += 1
print(cnt)
```

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from sklearn import tree
```

```python
target = final_df["Winner"]
```

```python
data = final_df.drop(columns=['Winner'],axis=1)
feature_names = data.columns
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

param_grid = {
    'n_estimators': [310,315,325],
    'criterion': ['gini','entropy']
}

grid = GridSearchCV(RandomForestClassifier(), param_grid, verbose=True, n_jobs=-1)

grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
```

```python
rf = RandomForestClassifier(n_estimators=315,criterion='entropy')
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
lr = make_pipeline(StandardScaler(),LogisticRegression(max_iter=1000, tol=1e-3))
lr = lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```
