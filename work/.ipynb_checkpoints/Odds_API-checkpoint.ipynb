```python
import pandas as pd
import requests
import json
import datetime as dt
from api_keys import api_key
```

```python
url = "https://api.the-odds-api.com/v3/odds/?"
sport = "mma_mixed_martial_arts"
query_url = f"{url}apiKey={api_key}&sport={sport}&region=us"
response = requests.get(query_url).json()
response
```

```python
def ml_api(data):
    fighter = []
    date = []
    h_odds = []
    a_odds = []
    
    for fight in range(len(data['data'])):
        event = data['data'][fight]
        fighter.append(event['teams'][0])
        fighter.append(event['teams'][1])
        time = dt.datetime.fromtimestamp(int(event['commence_time'])).strftime('%Y-%m-%d')
        date.append(time)
        date.append(time)
        h_odds_t = []
        a_odds_t = []
        for site in range(len(event['sites'])):
            h_odd = event['sites'][site]['odds']['h2h'][0]
            a_odd = event['sites'][site]['odds']['h2h'][1]  
            h_odds_t.append(h_odd)
            a_odds_t.append(a_odd)
        if event['sites_count'] == 0:
            h_odds.append('NaN')
            h_odds.append('NaN')
            a_odds.append('NaN')
            a_odds.append('NaN')
        else:
            h_odd = sum(h_odds_t)/len(h_odds_t)
            a_odd = sum(a_odds_t)/len(a_odds_t)
            if h_odd > 2:
                h_odd = round((h_odd - 1) * 100)
            elif h_odd == 1:
                h_odd = 'NaN'
            elif h_odd < 2:
                h_odd = round(100 / (1 - h_odd))
            else:
                h_odd = 100
            if a_odd > 2:
                a_odd = round((a_odd - 1) * 100)
            elif a_odd == 1:
                a_odd = 'NaN'
            elif a_odd < 2:
                a_odd = round(100 / (1 - a_odd))
            else:
                a_odd = 100
            h_odds.append(h_odd)
            h_odds.append(a_odd)
            a_odds.append(a_odd)
            a_odds.append(h_odd)
        
    game_dict = {
        "Fighter": fighter,
        "date": date,
        "H_odds": h_odds,
        "A_odds": a_odds
    }
    df = pd.DataFrame(game_dict)
    df = df.round({'H_odds':0,'A_odds':0})
    return df
```

```python
ml_df = ml_api(response)
ml_df
```

```python
ml_df.to_csv('../data/ml_input.csv', index=False)
```
